```python
from pyspark.sql.functions import col, lit
from pyspark.sql.types import StructType, StructField, StringType, TimestampType, DoubleType
import re

print('done')
```

```python
cities_dim = spark.read.parquet('/mnt/de-upskilling-weather/Silver/cities_dim.parquet/')
```

```python
# getting list of cities
cities_list = cities_dim.select("city").rdd.flatMap(lambda x: x).collect()
cities_list
```

```python
schema = StructType([
    StructField("city", StringType(), nullable=False),
    StructField("timestamp", TimestampType(), nullable=True),
    StructField("pred_temp", DoubleType(), nullable=True),
    StructField("country", StringType(), nullable=True),
    StructField("latitude", DoubleType(), nullable=True),
    StructField("longitude", DoubleType(), nullable=True),
    StructField("time_zone", StringType(), nullable=True)
])

# create empty forecast df
forecast_df = spark.createDataFrame(data=[], schema=schema)
```

```python
for city in cities_list:
    city_stripped = re.sub(r'\W+', '', city)
    
    # reading in forecast and joining with city_dim
    df = spark.read.parquet(f'/mnt/de-upskilling-weather/Silver/Forecasts/{city_stripped}_forecast.parquet/')
    df = df.withColumn('city', lit(city))
    df = df.join(cities_dim, on='city', how='left')
    
    forecast_df = forecast_df.union(df)
```

```python
forecast_df = forecast_df.orderBy('city', 'timestamp')
display(forecast_df)
```

```python
forecast_df.write.mode('overwrite').parquet('/mnt/de-upskilling-weather/Gold/temp_forecasts.parquet/')
```
